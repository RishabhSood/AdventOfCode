```python
import random

def colored_advent_tree(height=20):
    # Define ANSI colors
    COLORS = [
        "\033[31m",  # Red
        "\033[32m",  # Green
        "\033[33m",  # Yellow
        "\033[34m",  # Blue
        "\033[35m",  # Magenta
        "\033[36m"   # Cyan
    ]
    RESET = "\033[0m"  # Reset color

    # Characters to use for decorations
    decorations = ['@', '>', '<', '*', 'o', '0']

    # Tree topper
    print(" " * (height-1) + "\033[33m*\033[0m")  # Yellow star at the top

    # Build the tree
    for i in range(1, height + 1):
        spaces = " " * (height - i)  # Left padding
        branches = ""
        for _ in range(2 * i - 1):  # Generate random decorations
            char = random.choice(decorations)
            color = random.choice(COLORS)  # Pick a random color
            branches += f"{color}{char}{RESET}"
        print(spaces + branches)

    # Trunk
    for i in range(2):
        print(" " * (height - 1) + "\033[33m||\033[0m")  # Brown trunk
    print(" " * (height - 6) + "\033[90m- - - - - - -\033[0m")  # Gray base

# Generate the colored tree
colored_advent_tree(20)
```

Input

```python
file_path = "input.txt"  # Path to the input file

with open(file_path, "r") as f:
    data = f.read()

lock_n_keys = data.split('\n\n')

keys = set()
locks = set()

for elem in lock_n_keys:
    elem = elem.split('\n')
    if elem[0][0] == '.': #key
        rep_key = []
        for j in range(5):
            i = 6
            while i >= 0 and elem[i][j] == '#':
                i-=1
            rep_key.append(6 - (i+1))
        keys.add(tuple(rep_key))
    else: #lock
        rep_lock = []
        for j in range(5):
            i = 0
            while i < 7 and elem[i][j] == '#':
                i+=1
            rep_lock.append(i-1)
        locks.add(tuple(rep_lock))
```

Part 1

```python
cnt = 0
for lock in locks:
    for key in keys:
        flag = True
        for i in range(5):
            if key[i] + lock[i] > 5:
                flag = False
        if flag: cnt += 1
cnt
```
